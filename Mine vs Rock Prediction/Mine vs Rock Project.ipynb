```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
import joblib
```

### Exploaring the data

#### Read and Load the data

```python
df = pd.read_csv('Copy of sonar data.csv', header=None)
df.head()
```

```python
df.shape #Diminsions of the data (207 rows and 61 columns)
```

```python
df.tail()
```

```python
df.info()
```

### Data Preprocessing

```python
df.rename(columns={60 : 'Target'}, inplace=True)
```

```python
df.Target.value_counts() # 111 Mines and 96 Rocks
```

```python
df.describe()
```

```python
df.columns[df.isna().any()] # no null values
```

```python
df.groupby('Target').mean()
```

so it looks like that the coordinates of the Mine is always higher than the rock

```python
df.Target = df.Target.apply(lambda x: 1 if x == 'M' else 0)
df.Target.head()
```

```python
df.Target.tail()
```

### Assigning Feature and target variables

```python
X = df.drop('Target', axis=1)
y = df.Target
```

### PCA to extract important features and reduce diminishons

Use components such that 96% of variance is retaine

```python
pca = PCA(0.96)
X_pca = pca.fit_transform(X)
X_pca.shape
```

PCA returned 19 features from 61

### Split the data into Training and Testing Data

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca,y,test_size=0.25, random_state=195)
```

```python
len(X_train)
```

```python
len(X_test)
```

### Bulding and Training a Logistic Regression model

why Logistic Regression? after trying all the algorithms using GridSearchCV It turns out that Logistic Regression is the best

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

### accuracy of the model

```python
model.score(X_test, y_test)
```

### Save the model

```python
joblib.dump(model, 'Mine vs Rock Prediction')
```
